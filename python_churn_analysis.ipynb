## Churn by Monthly Charges and Total Charges

```python
# KDE plot for MonthlyCharges by Churn status
plot = sns.kdeplot(df_dummies['MonthlyCharges'][df_dummies['Churn'] == 0],
                   color='red', fill=True)

plot = sns.kdeplot(df_dummies['MonthlyCharges'][df_dummies['Churn'] == 1],
                   ax=plot, color='blue', fill=True)

plot.legend(['No Churn', 'Churn'], loc='upper right')
plot.set_ylabel('Density')
plot.set_xlabel('Monthly Charges')
plot.set_title('Monthly Charges by Churn')
plt.show()
```

```python
# Churn is high when the monthly charges are high. 
```

```python
# KDE plot for TotalCharges by Churn status
plot = sns.kdeplot(df_dummies['TotalCharges'][df_dummies['Churn'] == 0],
                   label='No Churn', color='red', fill=True)

plot = sns.kdeplot(df_dummies['TotalCharges'][df_dummies['Churn'] == 1],
                   label='Churn', ax=plot, color='blue', fill=True)

plot.legend(['No Churn', 'Churn'], loc='upper right')
plot.set_ylabel('Density')
plot.set_xlabel('Total Charges')
plot.set_title('Total Charges by Churn')
plt.show()
```

At first, it looks like customers with low total charges are more likely to leave.  
But this doesn't mean they spent less — it means they didn’t stay long.

Total charges depend on two things:
- Monthly charges
- How long the customer stayed (tenure)

So if someone has high monthly charges but only stayed for a short time,  
their total charges will still be low.

This means:
- High monthly charge
- Low tenure
- Low total charge  
→ All three are connected and lead to higher churn.

These are usually new customers who signed up for expensive plans but left early.
